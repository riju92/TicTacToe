```python
import numpy as np
dataset = np.loadtxt('./tictac_single.txt')
```

```python
np.random.shuffle(dataset)
```

```python
X = dataset[:, :9]
Y = dataset[:,9:]
```

```python
print(X.shape)
print(Y.shape)
```

```python
from sklearn.svm import LinearSVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC
```

```python
def linearSVM(train_X, test_X, train_Y, test_Y):
    print(test_X.shape)
    print(test_Y.shape)
    print(train_X.shape)
    print(train_Y.shape)
    #clf = make_pipeline(StandardScaler(), LinearSVC(tol=1e-3, multi_class = 'crammer_singer'))
    #clf.fit(train_X, train_Y)
    clf = OneVsRestClassifier(LinearSVC()).fit(train_X, train_Y)
    res = clf.predict(test_X)
    print(res)
    res_cm = res
    res = np.asarray(res)
    res = np.reshape(res, (-1, test_Y.shape[0]))
    accuracy = 0.0
    for i in range(res.shape[1]):
        if(res[0][i] == test_Y[i][0]):
            accuracy += 1
    print(accuracy)
    print("Accuracy = " + str(accuracy/res.shape[1]))
    print("#### CONFUSION MATRIX ##########")
    cm = confusion_matrix(test_Y, res_cm)
    cm = cm / cm.sum(axis=1)[:, np.newaxis] # normalizing the confusion matrix
    print(cm)
    return accuracy/res.shape[1], clf
```

```python
from sklearn.model_selection import StratifiedKFold
accuracy_max = 0.0
accuracy_min = 100.0
accuracy_avg = 0.0

best_clf = 0

k_cross_fold = 10

skf = StratifiedKFold(n_splits=k_cross_fold)
for train_index, test_index in skf.split(X, Y):
    #print(str(train_index[0]) + " and " + str(train_index[-1])  + " and " + "shape = " + str(train_index.shape))
    #print(str(test_index[0]) + " and " + str(test_index[-1])   + " and " + "shape = " + str(test_index.shape))
    train_X, test_X = X[train_index], X[test_index]
    train_Y, test_Y = Y[train_index], Y[test_index]
    
    accuracy, clf = linearSVM(train_X, test_X, train_Y, test_Y)
    accuracy *= 100.0
    
    if(accuracy > accuracy_max):
        accuracy_max = accuracy
        best_clf = clf
    accuracy_min = min(accuracy_min, accuracy)
    accuracy_avg += accuracy
    
    print("#######################")
    
print("Max accuracy = " + str(accuracy_max))
print("Min accuracy = " + str(accuracy_min))
print("Avg accuracy = " + str(accuracy_avg/k_cross_fold))

          
```
